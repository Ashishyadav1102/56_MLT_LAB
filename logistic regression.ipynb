```python
pd.get_dummies(titanic_data['Sex'])
#to convert the categorial column sex into numerical (dummy ) variable
```

```python
sex=pd.get_dummies(titanic_data['Sex'],drop_first=True)
sex.head(5)
```

```python
embark = pd.get_dummies(titanic_data['Embarked'],drop_first=True)
embark.head(5)
```

```python
pclass=pd.get_dummies(titanic_data['Pclass'],drop_first=True)
pclass.head(5)
```

```python
titanic_data=pd.concat([titanic_data,sex,embark,pclass],axis=1)
#to get back the coverted back
```

```python
titanic_data.head(5)
```

```python
titanic_data.drop(['Sex','Embarked','PassengerId','Name','Ticket','Pclass'],axis=1,inplace=True)
titanic_data.head(5)
```

```python
X=titanic_data.drop('Survived',axis=1)
# convert all column names to string to avoid typeErrror in scikit-learn
X.columns  = X.columns.astype(str)
y=titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
#
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```
